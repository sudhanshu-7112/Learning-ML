Data PreProcessing

```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
dataset = pd.read_csv('../Linear_Regression/Polynomial_Regression/position_salary.csv')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

Training Random Forest Regression on Dataset

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=77, random_state=0)
regressor.fit(x, y)
```

Predicting new Result

```python
regressor.predict([[6.5]])
```

Visualizing Random Forest Graph

```python
plt.scatter(x, y, color='red')
plt.plot(x, regressor.predict(x), color='blue')
plt.title('Random Forest')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
